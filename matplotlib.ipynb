Matplotlib:
  Useful for making plots

```python
#importing matplotlib library

import matplotlib.pyplot as plt

#import numpy to get data for out plots

import numpy as np
```

```python
x = np.linspace(0,10,100) # take 1000 evenly spaced value between 0 to 10

y = np.sin(x)

z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sine wave

plt.plot(x,y)

plt.xlabel("angle") # label of x axis

plt.ylabel("sine value")  # label of y axis

plt.title("sine wave") # title name

plt.show()
```

```python
#cosine wave

plt.plot(x,z)

plt.xlabel("angle")

plt.ylabel("cosine value")

plt.title("cosine wave")

plt.show()
```

```python
# parabola

x= np.linspace(-10,10,20)

y=x**2

plt.plot(x,y)

plt.show()
```
