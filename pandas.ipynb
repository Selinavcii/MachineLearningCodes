Pandas Dataframe:

Pandas DataFrame is two dimensional tabular data structure with labeled axes
( rows and columns )



```python
# importing the pandas library

import pandas as pd

import numpy as np
```

Creating a Pandas DataFrame



```python
from sklearn.datasets import fetch_california_housing

california_dataset = fetch_california_housing()

type(california_dataset)

print(california_dataset)
```

```python
# pandas dataFrame

california_df = pd.DataFrame(california_dataset.data,columns=california_dataset.feature_names)
```

```python
california_df.head() # take the first 5  row of the data
```

```python

california_df.shape
```
