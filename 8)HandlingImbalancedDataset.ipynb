Imbalanced Dataset:

a dataset with an unequal class distribution

```python
import numpy as np

import pandas as pd
```

```python
#loading the dataset to pandas dataframe

credit_card_data=pd.read_csv('/content/credit_data.csv')
```

```python
#fisrt 5 rows of df

credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#distribution of two casses

credit_card_data['Class'].value_counts()
```

The above result shows us imbalanced data distribution

0-->legit transactions
1--> fraudalent transactions

```python
#seperating the legit and fraudalent transactions

legit = credit_card_data[credit_card_data.Class==0]

fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)

print(fraud.shape)
```

under-sampling

building a sample dataset containing similar disribution of legit and fraudalent transactions

number of fraudalent transaction--> 108

```python
legit_sample = legit.sample(n=108)
```

```python
print(legit_sample.shape)
```

concatenate the two dataframe

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset['Class'].value_counts()
```
